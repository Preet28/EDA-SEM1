```python
df.isnull().mean()*100
df_missing_percent=100*df.isnull().sum()/len(df)
print('\nMissing percentage:\n\n', df_missing_percent)
```

```python
# Making a table of missing values with %
df_missing_table=pd.concat([df_missing, df_missing_percent], axis=1)
df_missing_table
```

```python
# Rename the columns
df_missing_table = df_missing_table.rename(columns = {0 : 'Missing Values', 1 : '% of Total Values'})
df_missing_table
```

 Using Missingno library

```python
import missingno as msno
```

Plotting bar plot

```python
msno.bar(df)
```

From above bar-plot, we can see the features 'ph', 'Sulfate', 'Trihalomethanes' have missing values.

<br>

On the left side of the plot, the scale of y-axis ranges from 0.0 to 1.0, where 1.0 represents 100% data completeness. <br><br>
If the bar is less than this, it depicts that we have missing values within that specific column. <br> <br>
On the right side of the plot, the scale is measured in index values. <br><br>
Top right represents the maximum number of rows within the dataframe. <br><br>
On the top of the plot, there are a series of numbers that represent the total count of the non-null values within that particular column.

--------

Plotting Matrix plot

```python
msno.matrix(df)
```

<h4>
