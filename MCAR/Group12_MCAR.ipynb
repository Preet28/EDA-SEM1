```python
from scipy.stats import chi2
```

```python
def little_mcar_test(data):
    n = len(data)
    groups = []
    for col in data.columns:
        mask = data[col].isnull()
        if mask.any():
            groups.append(mask.astype(int).values.reshape(-1,1))
    if len(groups) == 0:
        raise ValueError("No missing data found")
    r = np.concatenate(groups,axis=1)
    group_stats = r.T @ r
    m = len(groups)
    df = (n-1) * m
    chi2_stat = group_stats.trace()
    p_value = chi2.sf(chi2_stat,df)
    return {"chi2_stat":chi2_stat,"degree_of_freedom":df,"p_value":p_value}
```

```python
df = pd.read_csv('/content/EDA-SEM1/MCAR/NYC_Housing_Data.csv')
```

```python
result = little_mcar_test(df)
print(f"Chi-square statistic:{result['chi2_stat']}")
print(f"Degrees of freedom:{result['degree_of_freedom']}")
print(f"P-Value:{result['p_value']}")
```

Since the p-value is 1.0, which is significantly greater than the typical threshold of 0.05, the missing data in this dataset is likely Missing Completely at Random (MCAR). This suggests that the missing values are not dependent on the observed data or unobserved data, implying randomness in the missingness.
